**Detectar posible Diabetes**

Este es un problema simple de clasificación binaria para detectar la aparición de diabetes.

*Pasos a seguir:*

-Cargar datos.
-Definir el modelo.
-Compilar el modelo.
-Compatible con el modelo.
-Evaluar el modelo.
-Haga predicciones.

```python
# first neural network with keras tutorial
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## 1. Load Data

```python
df = pd.read_csv('data/pima-indians-diabetes.csv', header=None)
df.head()
```

All of the input variables that describe each patient are numerical. 

Hence, load the file as a matrix of numbers using the NumPy function `loadtxt()`.

There are eight input variables and one output variable (the last column). We will be learning a model to map rows of input variables (X) to an output variable (y), which we often summarize as `y = f(X)`

The variables can be summarized as follows:

<b>Input Variables (X):</b>

0. Número de veces embarazada
1. Concentración de glucosa plasmática a las 2 horas en una prueba de tolerancia a la glucosa oral.
2. Presión arterial diastólica (mm Hg)
3. Grosor del pliegue cutáneo del tríceps (mm)
4. Insulina sérica de 2 horas (mu U/ml)
5. Índice de masa corporal (peso en kg/(altura en m)^2)
6. Función del pedigrí de la diabetes
7. Edad (años)

<b>Output Variables (y):</b>

8. Variable de clasificación(0 or 1)


```python
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',  
         'BMI', 'DiabetesPedigreeFunction', 'Age', 'Class']
df.head()
```

### Method 2 to load data

```python
# load the dataset
dataset = loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
```

```python
dataset
```

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
X_scaled = scale(X) 
print('Scaled_X:\n', X_scaled)
```

```python
# Split dataset into 'train' & 'test' sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
# (optional): one hot encoding??
y_train = np_utils.to_categorical(y_train)

print('Y_Train Encoded:\n', y_train)
```

## 2. Definir Modelo Keras


The model design:

- 3 capas.
- 22 neuronas en total.
- Funciones de activación Relu y Sigmoide.

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

## 3. Compilar Modelo Keras


```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4. Ajustar Modelo Keras

```python
# fit the keras model on the dataset
history = model.fit(X_train, y_train, validation_split=0.33, epochs=150, batch_size=10)
# model.fit(X, y, epochs=150, batch_size=10, verbose=0)
```

## 5. Evaluación del Modelo

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

## 6. Predicciones

```python
# make class predictions with the model
predictions = np.argmax(model.predict(X_test), axis=-1)
# summarize the first 5 cases
for i in range(10):
    print('%s => %d (expected %d)' % (X_test[i].tolist(), predictions[i], y[i]))
```

```python
y_pred = model.predict(X_test)

y_pred = np.argmax(y_pred, axis=1) 

accuracy_score(y_test, y_pred)
```

```python
print(model.metrics_names)
```

Una de las devoluciones de llamada predeterminadas que se registra al entrenar todos los modelos de aprendizaje es la devolución de llamada Historial. Registra métricas de entrenamiento para cada época. Esto incluye la pérdida y la precisión (para problemas de clasificación), así como la pérdida y la precisión del conjunto de datos de validación, si se establece uno.

El objeto histórico se devuelve a partir de llamadas a la función fit() utilizada para entrenar el modelo. Las métricas se almacenan en un diccionario en el miembro del historial del objeto devuelto.

```python
print(history.history.keys())
```

## 7. Grafico de Perdida y precisión

Los gráficos pueden proporcionar una indicación de aspectos útiles sobre el entrenamiento del modelo, como por ejemplo:

- Es la velocidad de convergencia a lo largo de épocas (pendiente).
- Si el modelo ya ha convergido (meseta de la línea).
- Si el modo puede estar sobreaprendiendo los datos de entrenamiento (inflexión para la línea de validación).

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
