**Detectar posible Diabetes**

Este es un problema simple de clasificacin binaria para detectar la aparicin de diabetes.

*Pasos a seguir:*

-Cargar datos.
-Definir el modelo.
-Compilar el modelo.
-Compatible con el modelo.
-Evaluar el modelo.
-Haga predicciones.

```python
# first neural network with keras tutorial
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## 1. Load Data

```python
df = pd.read_csv('data/pima-indians-diabetes.csv', header=None)
df.head()
```

All of the input variables that describe each patient are numerical. 

Hence, load the file as a matrix of numbers using the NumPy function `loadtxt()`.

There are eight input variables and one output variable (the last column). We will be learning a model to map rows of input variables (X) to an output variable (y), which we often summarize as `y = f(X)`

The variables can be summarized as follows:

<b>Input Variables (X):</b>

0. Nmero de veces embarazada
1. Concentracin de glucosa plasmtica a las 2 horas en una prueba de tolerancia a la glucosa oral.
2. Presin arterial diastlica (mm Hg)
3. Grosor del pliegue cutneo del trceps (mm)
4. Insulina srica de 2 horas (mu U/ml)
5. ndice de masa corporal (peso en kg/(altura en m)^2)
6. Funcin del pedigr de la diabetes
7. Edad (aos)

<b>Output Variables (y):</b>

8. Variable de clasificacin(0 or 1)


```python
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',  
         'BMI', 'DiabetesPedigreeFunction', 'Age', 'Class']
df.head()
```

### Method 2 to load data

```python
# load the dataset
dataset = loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
```

```python
dataset
```

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
X_scaled = scale(X) 
print('Scaled_X:\n', X_scaled)
```

```python
# Split dataset into 'train' & 'test' sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
# (optional): one hot encoding??
y_train = np_utils.to_categorical(y_train)

print('Y_Train Encoded:\n', y_train)
```

## 2. Definir Modelo Keras


The model design:

- 3 capas.
- 22 neuronas en total.
- Funciones de activacin Relu y Sigmoide.

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

## 3. Compilar Modelo Keras


```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4. Ajustar Modelo Keras

```python
# fit the keras model on the dataset
history = model.fit(X_train, y_train, validation_split=0.33, epochs=150, batch_size=10)
# model.fit(X, y, epochs=150, batch_size=10, verbose=0)
```

## 5. Evaluacin del Modelo

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

## 6. Predicciones

```python
# make class predictions with the model
predictions = np.argmax(model.predict(X_test), axis=-1)
# summarize the first 5 cases
for i in range(10):
    print('%s => %d (expected %d)' % (X_test[i].tolist(), predictions[i], y[i]))
```

```python
y_pred = model.predict(X_test)

y_pred = np.argmax(y_pred, axis=1) 

accuracy_score(y_test, y_pred)
```

## 7. Grafico de Perdida y precisin

Los grficos pueden proporcionar una indicacin de aspectos tiles sobre el entrenamiento del modelo, como por ejemplo:

- Es la velocidad de convergencia a lo largo de pocas (pendiente).
- Si el modelo ya ha convergido (meseta de la lnea).
- Si el modo puede estar sobreaprendiendo los datos de entrenamiento (inflexin para la lnea de validacin).

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
